```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Penguins Dataset 

```python
penguins = sns.load_dataset("penguins")
```

```python
# First Preview of the dataset
print('Shape:', penguins.shape)
penguins.head()
```

```python
# Missing Values
penguins.isna().sum()
```

```python
# Missing Values with a Barplot
missing_values = penguins.isna().sum().reset_index()
missing_values.columns=['column', 'missing_values']

plt.figure(figsize=(8,4))
sns.barplot(data=missing_values, x='column', y='missing_values')
plt.title("Mising Values in The Dataset of Penguins")
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# A Histogram for the Body mass of the penguins
plt.figure(figsize=(8,6))
sns.histplot(data=penguins, x='body_mass_g', bins=30, color='burlywood')
plt.title('Penguins Body mass (g)')
plt.xlabel('Body Mass(g)')
plt.ylabel('Frequency')
plt.show()
```

### Data Cleaning

```python
# Dropping rows with missing data
clean_data = penguins.dropna(inplace=True)

print(f'Missing values after cleaning the data: {clean_data}')
```

```python
# Converting columns to the "category" type (from object)
category_columns = ['species', 'island', 'sex']
for col in category_columns:
    penguins[col] = penguins[col].astype('category')

print('Penguin columns data types:')
penguins.dtypes
```

```python
print("Data after Cleaning")
penguins.info()
```

```python
# Histogram for numeric Columns using KDE
numeric_columns=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

plt.figure(figsize=(10,8))
for i, col in enumerate(numeric_columns):
    plt.subplot(2, 2, i+1)
    sns.histplot(penguins[col], kde=True, bins=20, edgecolor='black', linewidth=1.2)
    plt.title(col)

plt.tight_layout()
plt.show()
```

## Data Transformation

```python
# Creating a new column for the body mass of the penguins(in kg)
penguins["body_mass_kg"] = penguins["body_mass_g"] / 1000
```

```python
# Bill length to depth ratio
# Comparing how long and deep a penguin's bill is (A higher ratio means linger/thinner bill ; lower ratio -> shorter/deeper bill)
penguins["bill_ratio"] = penguins["bill_length_mm"] / penguins["bill_depth_mm"]
print(f"Ratio: {penguins['bill_ratio']}")
```

```python
# Grouping flipper length into categories (size groups)
penguins["flipper_size"] = pd.cut(
    penguins["flipper_length_mm"],
    bins=[170, 190, 210, 230],
    labels=["Short", "Medium", "Long"]
)
```

```python
penguins.head()
```

### Data Summarization

```python
penguins_clean = penguins.copy()

penguins_clean['bill_area_mm2'] = penguins_clean['bill_length_mm'] * penguins_clean['bill_depth_mm']

# BMI-like metric (body mass / flipper length)
penguins_clean['mass_flipper_ratio'] = penguins_clean['body_mass_g'] / penguins_clean['flipper_length_mm']

penguins_clean = penguins_clean.dropna()
species_summary = penguins_clean.groupby('species', observed=True).agg({
    'bill_length_mm': ['mean', 'std'],
    'bill_depth_mm': ['mean', 'std'],
    'flipper_length_mm': ['mean', 'std'],
    'mass_flipper_ratio': ['mean', 'std'],
    'bill_area_mm2': 'mean'
}).round(2)

print("Statistics by species:")
species_summary
```

```python
mass_by_species_sex = pd.pivot_table(penguins_clean, observed=True,
                                   values='body_mass_g', 
                                   index='species', 
                                   columns='sex', 
                                   aggfunc='mean').round(1)

print("Average body mass (g) by species and sex:")
mass_by_species_sex
```

```python
key_insights_table = penguins_clean.groupby('species', observed=True).agg({
    'body_mass_g': 'mean',
    'flipper_length_mm': 'mean',
    'bill_length_mm': 'mean',
    'bill_area_mm2': 'mean'
}).round(2)

print("Morphological differences by species:")
key_insights_table
```

### Visualization

```python
# Plot 1: Distribution of body mass by species
plt.figure(figsize=(12, 6))
sns.boxplot(data=penguins_clean, x='species', y='body_mass_g', hue='sex')
plt.title('Distribution of body mass by species and sex')
plt.xlabel('Penguin Species')
plt.ylabel('Body Mass (g)')
plt.legend(title='Sex')
plt.show()

print("Result: Male penguins are generally larger than female penguins.")
```

```python
# Plot 2: Relationship between bill dimensions by species
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=penguins_clean, 
    x='bill_length_mm', 
    y='bill_depth_mm', 
    hue='species', 
    style='species', 
    s=100
)
plt.title('Bill dimensions by penguin species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Result: Gentoo penguins generally have longer bills, whereas Adelie penguins have deeper bills.")
```

```python
# Plot 3: Flipper length by species and island
plt.figure(figsize=(12, 6))
sns.violinplot(
    data=penguins_clean, 
    x='species', 
    y='flipper_length_mm', 
    hue='island', 
    split=True
)
plt.title('Flipper length by species and island')
plt.xlabel('Penguin Species')
plt.ylabel('Flipper Length (mm)')
plt.legend(title='Island', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Result: Gentoo penguins have longer flippers than other species, with less variation across different islands.")
```

```python
# Plot 4: Correlation heatmap of morphological features
plt.figure(figsize=(8, 6))
numerical_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
correlation_matrix = penguins_clean[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlation among penguin morphological features')
plt.tight_layout()
plt.show()

print("Result: Positive correlation between body mass, flipper length, and bill length, indicating that these features tend to scale together.")
```

```python
# Plot 5: Body mass vs. flipper length by species
plt.figure(figsize=(10, 6))
sns.lmplot(
    data=penguins_clean, 
    x='flipper_length_mm', 
    y='body_mass_g', 
    hue='species', 
    height=6, 
    aspect=1.2
)
plt.title('Body mass vs. flipper length by species')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.tight_layout()
plt.show()

print("Result: Positive relationship between flipper length and body mass for each species, with Gentoo penguins showing the largest size.")
```
