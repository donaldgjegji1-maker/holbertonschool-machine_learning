```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Admin\Downloads\albania_house_prices.csv')
print(df.shape)
```

```python
property_counts=df['Property_Type'].value_counts()
```

```python
plt.figure(figsize=(8,6))
```

```python
plt.pie(property_counts, 
        explode=(0.05, 0,0 ), autopct='%1.1f%%',
        labels=property_counts.index, 
        colors=['skyblue', 'salmon', 'lightgreen'])
plt.show()
```
