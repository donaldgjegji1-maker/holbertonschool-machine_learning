```python
# Statistics information for the dataset
penguins.describe()
```

```python
# Missing Values
penguins.isna().sum()
```

```python
# Missing Values with a Barplot
missing_values = penguins.isna().sum().reset_index()
missing_values.columns=['column', 'missing_values']

plt.figure(figsize=(8,4))
sns.barplot(data=missing_values, x='column', y='missing_values')
plt.title("Mising Values in The Dataset of Penguins")
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.xticks(rotation=45)
plt.show()
```

```python
# Counting Unique species
penguins['species'].value_counts()
```

```python
# Countplot graphic for uniques species count
plt.figure(figsize=(8,6))
sns.countplot(data=penguins, x='species')
plt.title('Penguin species count')
plt.xlabel('Species')
plt.ylabel('Counts')
plt.show()
```

```python
# A Histogram for the Body mass of the penguins
plt.figure(figsize=(8,6))
sns.histplot(data=penguins, x='body_mass_g', bins=30, color='burlywood')
plt.title('Penguins Body mass (g)')
plt.xlabel('Body Mass(g)')
plt.ylabel('Frequency')
plt.show()
```

### Data Cleaning

```python
# Dropping rows with missing data
clean_data = penguins.dropna(inplace=True)

print(f'Missing values after cleaning the data: {clean_data}')
```

```python
# Converting columns to the "category" type (from object)
category_columns = ['species', 'island', 'sex']
for col in category_columns:
    penguins[col] = penguins[col].astype('category')

print('Penguin columns data types:')
penguins.dtypes
```

```python
print("Data after Cleaning")
penguins.info()
```

```python
# Histogram for numeric Columns using KDE
numeric_columns=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

plt.figure(figsize=(10,8))
for i, col in enumerate(numeric_columns):
    plt.subplot(2, 2, i+1)
    sns.histplot(penguins[col], kde=True, bins=20, edgecolor='black', linewidth=1.2)
    plt.title(col)

plt.tight_layout()
plt.show()
```

## Data Transformation

```python
# Creating a new column for the body mass of the penguins(in kg)
penguins["body_mass_kg"] = penguins["body_mass_g"] / 1000
```

```python
# Bill length to depth ratio
# Comparing how long and deep a penguin's bill is (A higher ratio means linger/thinner bill ; lower ratio -> shorter/deeper bill)
penguins["bill_ratio"] = penguins["bill_length_mm"] / penguins["bill_depth_mm"]
print(f"Ratio: {penguins['bill_ratio']}")
```

```python
# Grouping flipper length into categories (size groups)
penguins["flipper_size"] = pd.cut(
    penguins["flipper_length_mm"],
    bins=[170, 190, 210, 230],
    labels=["Short", "Medium", "Long"]
)
```

```python
penguins.head()
```

### Data Summarization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins_clean = penguins.copy()

penguins_clean['siperfaqja_sqepit_mm2'] = penguins_clean['bill_length_mm'] * penguins_clean['bill_depth_mm']

# Metrika BMI-like (body mass / flipper length)
penguins_clean['raporti_mase_pende'] = penguins_clean['body_mass_g'] / penguins_clean['flipper_length_mm']

penguins_clean = penguins_clean.dropna()
species_summary = penguins_clean.groupby('species').agg({
    'bill_length_mm': ['mean', 'std'],
    'bill_depth_mm': ['mean', 'std'],
    'flipper_length_mm': ['mean', 'std'],
    'raporti_mase_pende': ['mean', 'std'],
    'siperfaqja_sqepit_mm2': 'mean'
}).round(2)

print("Statistika per specie:")
species_summary
```

```python
mass_by_species_sex = pd.pivot_table(penguins_clean, 
                                   values='body_mass_g', 
                                   index='species', 
                                   columns='sex', 
                                   aggfunc='mean').round(1)

print("Pesha (g) mesatare per specie dhe gjini:")
mass_by_species_sex
```

```python
key_insights_table = penguins_clean.groupby('species').agg({
    'body_mass_g': 'mean',
    'flipper_length_mm': 'mean',
    'bill_length_mm': 'mean',
    'siperfaqja_sqepit_mm2': 'mean'
}).round(2)

print("Diferenca morfologjike per specie:")
key_insights_table
```

### Visualization

```python
# Plot 1: Shperndarja e peshes ne varesi te species
plt.figure(figsize=(12, 6))
sns.boxplot(data=penguins_clean, x='species', y='body_mass_g', hue='sex')
plt.title('Shperndarja e peshes per specie dhe gjini')
plt.xlabel('Specie Pinguin')
plt.ylabel('Pesha (g)')
plt.legend(title='Gjinia')
plt.show()

print("Rezultati: Pinguinet meshkuj jane zakonisht me te medhenj se pinguinet femra")
```

```python
# Plot 2: Lidhja mes dimensioneve te sqepit per specie
plt.figure(figsize=(10, 6))
sns.scatterplot(data=penguins_clean, x='bill_length_mm', y='bill_depth_mm', 
                hue='species', style='species', s=100)
plt.title('Dimensionet e sqepit per specie pinguin')
plt.xlabel('Gjatesia e sqepit (mm)')
plt.ylabel('Thellesia e sqepit (mm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Rezultati: Pinguinet Gentoo zakonisht kane sqep me te gjate ndersa pinguinet Adelie kane sqep me te thelle")
```

```python
# Plot 3: Gjatesia e pendeve per specie dhe ishull
plt.figure(figsize=(12, 6))
sns.violinplot(data=penguins_clean, x='species', y='flipper_length_mm', hue='island', split=True)
plt.title('Gjatesia e pendeve per species dhe ishull')
plt.xlabel('Specie Pinguin')
plt.ylabel('Gjatesia e pendeve (mm)')
plt.legend(title='Ishulli', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("Rezultati: Pinguinet Gentoo kane pende me te gjata se speciet tjera, me pak variacion ne ishuj te ndryshem")
```

```python
# Plot 4: Correlation heatmap per "veori" te nbdryshme morfologjike
plt.figure(figsize=(8, 6))
numerical_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
correlation_matrix = penguins_clean[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Korrelacioni mes "veorive" morfologjike te pinguineve')
plt.tight_layout()
plt.show()

print("Rezultati: Korrelacion pozitiv mes peshes, gjatesise se pendeve dhe gjatesis se sqepit, cka na thote qe keto veori behen 'scale' se bashku")
```

```python
# Plot 5: Peshe vs Gjatesi pende per specie
plt.figure(figsize=(10, 6))
sns.lmplot(data=penguins_clean, x='flipper_length_mm', y='body_mass_g', 
           hue='species', height=6, aspect=1.2)
plt.title('Peshe vs Gjatesi pende per specie')
plt.xlabel('Gjatesia e pendeve (mm)')
plt.ylabel('Pesha (g)')
plt.tight_layout()
plt.show()

print("Rezultati: Marredhenie pozitive midis gjatesise se pendeve dhe mases trupore per secilen specie, me pinguinet Gentoo q tregojn madhesin me te madhe")
```
