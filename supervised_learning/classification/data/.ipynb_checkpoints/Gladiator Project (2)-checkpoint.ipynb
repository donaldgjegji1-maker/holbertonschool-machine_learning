## : Feature Engineering




### Age
Ages range from young fighters in their teens to veterans in their 50s. Continuous age data makes it hard to analyze performance patterns across different life stages.

Let's group into combat-relevant age brackets:

- Teen (0-20) - Developing skills, high agility
- Young Adult (21-25) - Peak physical condition
- Adult (26-30) - Experience building
- Prime (31-35) - Peak performance years
- Mature (36-40) - Veteran wisdom
- Veteran (41-50) - Seasoned fighters
- Elder (51+) - Rare survivors

```python
print("\n1. CREATING AGE GROUPS")
gladis['Age_Group'] = pd.cut(gladis['Age'], 
                            bins=[0, 20, 25, 30, 35, 40, 50, 100], 
                            labels=['Teen', 'Young Adult', 'Adult', 'Prime', 'Mature', 'Veteran', 'Elder'])
print("Age group distribution:")
print(gladis['Age_Group'].value_counts().sort_index())
```

### Height
Heights range from very short to exceptionally tall fighters. Height affects reach, intimidation factor, and fighting style effectiveness.

Let's create standardized height tiers:

- Short (0-165cm) - Close-quarters advantage
- Average (166-175cm) - Balanced fighters
- Tall (176-185cm) - Reach advantage
- Very Tall (186-195cm) - Significant reach
- Giant (196cm+) - Intimidation factor
  

```python
print("\n2. CREATING HEIGHT CATEGORIES")
gladis['Height_Category'] = pd.cut(gladis['Height'],
                                  bins=[0, 165, 175, 185, 195, 300],
                                  labels=['Short', 'Average', 'Tall', 'Very Tall', 'Giant'])
print("Height category distribution:")
print(gladis['Height_Category'].value_counts().sort_index())
```

### Weight

Weights range from lightweight to super heavyweight fighters. Weight impacts power, endurance, and suitable fighting styles.

Let's implement combat sports weight classes:

- Flyweight (0-60kg) - Speed and agility
- Lightweight (61-70kg) - Balanced speed/power
- Welterweight (71-80kg) - Athletic build
- Middleweight (81-90kg) - Power fighters
- Heavyweight (91-100kg) - Strong contenders
- Super Heavyweight (101kg+) - Raw power

```python
print("\n3. CREATING WEIGHT CLASSES")
gladis['Weight_Class'] = pd.cut(gladis['Weight'],
                               bins=[0, 60, 70, 80, 90, 100, 150],
                               labels=['Flyweight', 'Lightweight', 'Welterweight', 'Middleweight', 'Heavyweight', 'Super Heavyweight'])
print("Weight class distribution:")
print(gladis['Weight_Class'].value_counts().sort_index())
```

### BMI Categories - Body Composition Analysis
BMI ranges from underweight to severely obese. Body composition affects endurance, mobility, and combat effectiveness.

Standard health BMI categories adapted for combat:

- Underweight (<18.5) - Potential strength issues
- Normal (18.5-22.9) - Optimal conditioning
- Overweight (23-24.9) - Power build
- Obese I (25-29.9) - Strength focus
- Obese II (30-34.9) - Endurance concerns
- Obese III (35+) - Mobility challenges

```python
print("\n4. CREATING BMI CATEGORIES")
gladis['BMI'] = gladis['Weight'] / ((gladis['Height'] / 100) ** 2)
gladis['BMI_Category'] = pd.cut(gladis['BMI'],
                               bins=[0, 18.5, 22.9, 24.9, 29.9, 34.9, 100],
                               labels=['Underweight', 'Normal', 'Overweight', 'Obese I', 'Obese II', 'Obese III'])
print("BMI category distribution:")
print(gladis['BMI_Category'].value_counts().sort_index())
```

### Experience Levels
Battle experience ranges from 0 to hundreds of fights. Let's categorize experience levels, as well, into these
experience brackets:

- Rookie (0-5 fights) - Learning phase
- Experienced (6-20 fights) - Developing skills
- Veteran (21-50 fights) - Battle-tested
- Elite (51-100 fights) - Master fighters
- Legend (101+ fights) - Rare survivors

```python
print("\n5. CREATING EXPERIENCE LEVELS")
gladis['Experience_Level'] = pd.cut(gladis['Battle Experience'],
                                   bins=[0, 5, 20, 50, 100, 1000],
                                   labels=['Rookie', 'Experienced', 'Veteran', 'Elite', 'Legend'])
print("Experience level distribution:")
print(gladis['Experience_Level'].value_counts().sort_index())
```

### Performance
Win ratios range from 0% to 100% success rates. Continuous ratios make it hard to compare performance tiers, so let's have some 
clear performance categories:

- Poor (0-20%) - Struggling fighters
- Below Average (21-40%) - Developing skills
- Average (41-60%) - Competitive fighters
- Good (61-80%) - Strong performers
- Excellent (81-100%) - Elite champions

```python
print("\n6. CREATING PERFORMANCE METRICS")
gladis['Win_Ratio'] = gladis['Wins'] / (gladis['Wins'] + gladis['Losses'])
gladis['Win_Ratio'] = gladis['Win_Ratio'].fillna(0)

gladis['Performance_Tier'] = pd.cut(gladis['Win_Ratio'],
                                   bins=[0, 0.2, 0.4, 0.6, 0.8, 1.01],
                                   labels=['Poor', 'Below Average', 'Average', 'Good', 'Excellent'])
print("Performance tier distribution:")
print(gladis['Performance_Tier'].value_counts().sort_index())
```

### Popularity 
Public favor scores range from 0 to 1. Continuous popularity metrics obscure crowd appeal categories.

Solution: Audience perception tiers:

- Hated (0-0.2) - Crowd disapproval
- Unpopular (0.21-0.4) - Limited support
- Neutral (0.41-0.6) - Average popularity
- Popular (0.61-0.8) - Crowd favorites
- Celebrity (0.81-1.0) - Arena stars

```python
print("\n7. CREATING PUBLIC FAVOR LEVELS")
gladis['Public_Favor_Level'] = pd.cut(gladis['Public Favor'],
                                     bins=[0, 0.2, 0.4, 0.6, 0.8, 1.01],
                                     labels=['Hated', 'Unpopular', 'Neutral', 'Popular', 'Celebrity'])
print("Public favor level distribution:")
print(gladis['Public_Favor_Level'].value_counts().sort_index())
```

## Survival rate

### Understanding Gladiator Survival Patterns

This analysis examines survival rates across different demographic and performance factors to identify what characteristics contribute to a gladiator's longevity in the arena. Survival is the ultimate measure of success - understanding these patterns can reveal which attributes help fighters survive their brutal profession.

The following visualizations break down survival rates by:

- Age Groups: How different life stages affect survival chances
- Experience Levels: Whether battle-hardened veterans outlast newcomers
- Physical Condition: How BMI categories correlate with survival
- Public Perception: Whether crowd favor translates to longer careers
Each plot shows precise survival rates with reference lines indicating the overall average, allowing for clear comparison across categories.

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 12))

overall_survival = gladis['Survived'].mean()

# Define all plots in one structure
plot_settings = [
    ("Age_Group", "Survival Rate by Age Group",
     ['#ff9999', '#ffb366', '#ffcc33', '#66b3ff', '#3366cc', '#9933cc'], axes[0, 0]),

    ("Experience_Level", "Survival Rate by Experience Level",
     ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc'], axes[0, 1]),

    ("BMI_Category", "Survival Rate by BMI Category",
     ['#ff6b6b', '#4ecdc4', '#45b7d1', '#96ceb4', '#feca57', '#ff9ff3'], axes[1, 0]),

    ("Public_Favor_Level", "Survival Rate by Public Favor Level",
     ['#ff6b6b', '#ffa726', '#ffee58', '#9ccc65', '#26a69a'], axes[1, 1])
]

# Generic function that handles all plots
def plot_survival(ax, group_col, title, colors):
    series = gladis.groupby(group_col, observed=False)['Survived'].mean()
    bars = ax.bar(series.index, series.values,
                  color=colors[:len(series)], alpha=0.8, edgecolor='black', linewidth=0.5)

    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.grid(axis='y', alpha=0.3)

    # dynamic y-limits
    mn, mx = series.min(), series.max()
    r = mx - mn
    ax.set_ylim(mn - r * 0.1, mx + r * 0.15)

    # value labels
    for bar, value in zip(bars, series.values):
        ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + r * 0.01,
                f"{value:.3f}", ha="center", va="bottom", fontweight="bold",
                bbox=dict(boxstyle="round,pad=0.2", facecolor="white", alpha=0.8))

    # overall line
    ax.axhline(y=overall_survival, color="red", linestyle="--", alpha=0.7,
               linewidth=1, label=f"Overall: {overall_survival:.3f}")
    ax.legend()

    # special case: rotate Public Favor axes
    if group_col == "Public_Favor_Level":
        ax.tick_params(axis="x", rotation=45)

# Loop through everything
for group_col, title, colors, ax in plot_settings:
    plot_survival(ax, group_col, title, colors)

plt.tight_layout()
plt.show()
```

### Measuring Gladiator Effectiveness in Battle

While survival is crucial, combat performance tells us who truly dominates in the arena. This analysis focuses on winning patterns to identify the most effective weapons and the most successful origins for producing champion gladiators.

Key questions explored:

- Weapon Effectiveness: Which arms yield the highest victory counts?
- Regional Excellence: Which origins produce the most successful fighters?

The visualizations provide detailed comparisons of average wins across different categories, highlighting even subtle differences in performance to identify true competitive advantages.

```python
plt.figure(figsize=(16, 8))

overall_avg = gladis['Wins'].mean()

# Generic plotting function
def plot_avg_bar(data, title, colors, subplot):
    plt.subplot(1, 2, subplot)

    mn, mx = data.min(), data.max()
    r = mx - mn
    y_min, y_max = mn - r * 0.1, mx + r * 0.2

    bars = plt.bar(data.index, data.values,
                   color=colors[:len(data)], alpha=0.8,
                   edgecolor="black", linewidth=0.5)

    plt.title(title, fontsize=14, fontweight="bold")
    plt.ylabel("Average Wins")
    plt.xticks(rotation=45, ha="right")
    plt.ylim(y_min, y_max)
    plt.grid(axis="y", alpha=0.3)

    # Value labels
    for bar, val in zip(bars, data.values):
        plt.text(bar.get_x() + bar.get_width()/2,
                 bar.get_height() + r * 0.01,
                 f"{val:.2f}", ha="center", va="bottom",
                 fontweight="bold", fontsize=8,
                 bbox=dict(boxstyle="round,pad=0.2",
                           facecolor="white", alpha=0.8))

    # Overall average line
    plt.axhline(y=overall_avg, color="red",
                linestyle="--", alpha=0.7, linewidth=1,
                label=f"Overall Avg: {overall_avg:.2f}")
    plt.legend()


# --- Plot 1: Weapons ---
weapon_wins = (
    gladis.groupby('Weapon of Choice', observed=False)['Wins']
    .mean().sort_values(ascending=False).head(15)
)
weapon_colors = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#3498db', '#9b59b6']
plot_avg_bar(weapon_wins, "Average Wins by Weapon\n(Most Effective Weapons)", weapon_colors, 1)



# --- Plot 2: Origins ---
origin_wins = (
    gladis.groupby('Origin', observed=False)['Wins']
    .mean().sort_values(ascending=False)
)
origin_colors = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#3498db', '#9b59b6']
plot_avg_bar(origin_wins, "Average Wins by Origin\n(All 6 Origins)", origin_colors, 2)


plt.tight_layout()
plt.show()
```

### Understanding Relationships Between Gladiator Attributes
This correlation matrix reveals how different gladiator characteristics interact with each other, helping us identify which factors are most strongly associated with success in the arena. Correlation values range from -1 (perfect negative relationship) to +1 (perfect positive relationship), with 0 indicating no linear relationship.

Key relationships to examine:

- Performance Drivers: Which attributes most strongly correlate with wins and performance scores?
- Physical Advantages: How do height, weight, and BMI relate to combat success?
- Psychological Factors: The impact of mental resilience and public favor on outcomes
- Experience Benefits: How battle experience translates to improved performance

The heatmap uses a color scale where:

 Red shades indicate positive correlations
 Blue shades indicate negative correlations
 White/light colors show weak or no correlation

```python
gladis['Performance_Score'] = (
    gladis['Wins'] * 0.6 +
    gladis['Win_Ratio'] * 0.3 +
    gladis['Public Favor'] * 0.1
)

numerical_cols = ['Age', 'Height', 'Weight', 'Wins', 'Losses', 'Public Favor', 
                  'Mental Resilience', 'Battle Experience', 'BMI', 'Win_Ratio', 'Performance_Score']

correlation_matrix = gladis[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```

### : Results: Which gladiators had the better chances to win their fights

After analyzing 9,976 gladiators, their physical attributes, psychological traits, experience, weapons, strategies, social background, and battle outcomes, a clear picture emerges about what truly determined success in the Roman arena.

Gladiatorial combat was never random, nor was it decided purely by brute strength.
Instead, victory depended on a complex combination of skill, preparation, social backing, physical conditioning, and tactical decision-making.
Our analysis shows that the gladiators with the highest chances to win (and survive) shared several consistent characteristics.


* Physical Strength Matters  But Only in Balance

Tall or heavy fighters did not automatically win more.
The most successful gladiators fell into balanced height and weight ranges, especially:

Height: 170185 cm

Weight: 7595 kg

BMI: Normal  Overweight (ideal conditioning)

These fighters had enough muscle mass for power, but not so much that it slowed them down.

* Experience Was One of the Strongest Predictors of Success

Veteran fighters consistently outperformed rookies.

Battle Experience strongly correlated with wins (0.95)

Veteran & Prime-age fighters had the highest win & survival rates

Rookies had the lowest

* Fighting Strategy Was a Critical Survival Factor

Gladiators who fought aggressively or balanced had higher survival rates than defensive fighters.

Aggressive  Highest survival (54.6%)

Balanced  Solid performance

Defensive  Lowest survival (33.4%)

This aligns with historical accounts: defensive fighters risked being cornered or overwhelmed.


> **Gladiator Winning Profile (The Most Successful Type of Fighter)**

A gladiator had the best chance to win if he was:

**Physical Traits**
* Height: 170185 cm
* Weight: 7595 kg
* BMI: Normal to slightly Overweight (strong but agile)

**Combat Ability**
* Veteran or Experienced fighter
* Chose high-impact weapons (Sica, Trident, Mace)
* Fought with Aggressive or Balanced strategy

**Psychological & Social Factors**
* High Mental Resilience
* Popular with the crowd
* Backed by wealthy patrons
* Equipped with superior gear
* Member of strong Allegiance Networks

**Background**
* Origin regions such as Rome and Numidia produced slightly higher-performing gladiators.
